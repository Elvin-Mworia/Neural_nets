```python
# Imports math library
import numpy as np
# Imports plotting library
import matplotlib.pyplot as plt
```

```python
# Define a linear function with just one input, x
def linear_function_1D(x,beta,omega):
  # TODO -- replace the code line below with formula for 1D linear equation
  y = omega* x + beta

  return y
```

```python
# Plot the 1D linear function

# Define an array of x values from 0 to 10 with increments of 0.01
# https://numpy.org/doc/stable/reference/generated/numpy.arange.html
x = np.arange(0.0,10.0, 0.01)
# Compute y using the function you filled in above
beta = -2.25; omega = 10.2

y = linear_function_1D(x,beta,omega)

# Plot this function
fig, ax = plt.subplots()
ax.plot(x,y,'r-')
ax.set_ylim([0,10]);ax.set_xlim([0,10])
ax.set_xlabel('x'); ax.set_ylabel('y')
plt.show

# TODO -- experiment with changing the values of beta and omega
# to understand what they do.  Try to make a line
# that crosses the y-axis at y=10 and the x-axis at x=5
```

```python
# Code to draw 2D function -- read it so you know what is going on, but you don't have to change it
def draw_2D_function(x1_mesh, x2_mesh, y):
    fig, ax = plt.subplots()
    fig.set_size_inches(7,7)
    pos = ax.contourf(x1_mesh, x2_mesh, y, levels=256 ,cmap = 'hot', vmin=-10,vmax=10.0)
    fig.colorbar(pos, ax=ax)
    ax.set_xlabel('x1');ax.set_ylabel('x2')
    levels = np.arange(-10,10,1.0)
    ax.contour(x1_mesh, x2_mesh, y, levels, cmap='winter')
    plt.show()
```

```python
def linear_function_3D(x1, x2, x3, omega1, omega2, omega3, beta):
  # TODO -- replace the code below with formula for a single 3D linear equation
    y = omega1 * x1 + omega2 * x2 + omega3 * x3 + beta
    return y
```

```python
# Define the parameters
beta1 = 0.5; beta2 = 0.2
omega11 =  -1.0 ; omega12 = 0.4; omega13 = -0.3
omega21 =  0.1  ; omega22 = 0.1; omega23 = 1.2

# Define the inputs
x1 = 4 ; x2 =-1; x3 = 2

# Compute using the individual equations
y1 = linear_function_3D(x1,x2,x3,beta1,omega11,omega12,omega13)
y2 = linear_function_3D(x1,x2,x3,beta2,omega21,omega22,omega23)
print("Individual equations")
print('y1 = %3.3f\ny2 = %3.3f'%((y1,y2)))

# Define vectors and matrices
beta_vec = np.array([[beta1],[beta2]])
omega_mat = np.array([[omega11,omega12,omega13],[omega21,omega22,omega23]])
x_vec = np.array([[x1], [x2], [x3]])

# Compute with vector/matrix form
y_vec = beta_vec+np.matmul(omega_mat, x_vec)
print("Matrix/vector form")
print('y1= %3.3f\ny2 = %3.3f'%((y_vec[0],y_vec[1])))
```


2.  Write code to compute three linear equations with two inputs ($x_1$, $x_2$) using both the individual equations and the matrix form (you can make up any values for the inputs $\beta_{i}$ and the slopes $\omega_{ij}$.


 three linear equations are defined with two inputs (x1, x2) as follows:

y1 = a11x1 + a12x2 + b1
y2 = a21x1 + a22x2 + b2
y3 = a31x1 + a32x2 + b3

The  equations will be computed individually and in matrix form


```python

# Define the slopes (a) and intercepts (b)
a11, a12, b1 = 1, 2, 3
a21, a22, b2 = 4, 5, 6
a31, a32, b3 = 7, 8, 9

# Define the inputs (x)
x1, x2 = 10, 11

# Compute the equations individually
y1 = a11*x1 + a12*x2 + b1
y2 = a21*x1 + a22*x2 + b2
y3 = a31*x1 + a32*x2 + b3

print(f"Individual computations: y1={y1}, y2={y2}, y3={y3}")

# Compute the equations in matrix form
M = np.array([[a11, a12], [a21, a22], [a31, a32]])  # slopes
x = np.array([x1, x2])  # inputs
b = np.array([b1, b2, b3])  # intercepts

y = np.dot(M, x) + b

print(f"Matrix computation: y={y}")
```

5. A function is convex if we can draw a straight line between any two points on the function, and this line always lies above the function. Similarly, a function is concave if a straight line between any two points always lies below the function.  Is the exponential function convex or concave or neither?


#Answers
1.exp(0): Any number raised to the power of 0 is 1, so exp(0) = 1.

2.exp(1): The value of e raised to the power of 1 is e itself, so exp(1) = e ≈ 2.71828.

3.exp(-∞): As x approaches negative infinity, exp(x) approaches 0, so exp(-∞) = 0.

4.exp(+∞): As x approaches positive infinity, exp(x) also approaches positive infinity, so exp(+∞) = +∞

5.The exponential function, exp(x), is convex. This means that for any two points on the function, the line segment connecting them lies above the function. This property holds because the second derivative of the exponential function is always positive

In mathematical terms, if f(x) is the exponential function, then the second derivative f''(x) = exp(x), which is always greater than 0 for all real x. Therefore, the exponential function is convex
